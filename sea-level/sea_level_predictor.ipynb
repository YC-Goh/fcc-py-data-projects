```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
fig, ax = plt.subplots(1,1)
ax.plot('Year', 'CSIRO Adjusted Sea Level', '.k', data=df)
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
```

```python
ols_1 = linregress(df[['Year', 'CSIRO Adjusted Sea Level']])
range_1 = pd.RangeIndex(start=df['Year'].min(), stop=2050, step=1)
ax.plot(range_1, range_1 * ols_1.slope + ols_1.intercept, '-r')
fig
```

```python
ols_2 = linregress(df[df['Year']>=2000][['Year', 'CSIRO Adjusted Sea Level']])
range_2 = pd.RangeIndex(start=2000, stop=2050, step=1)
ax.plot(range_2, range_2 * ols_2.slope + ols_2.intercept, '-b')
fig
```
